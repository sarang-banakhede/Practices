```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
x,y=make_classification(n_features=2,n_informative=1,n_classes=2,n_clusters_per_class=1,n_samples=200,n_redundant=0,random_state=41,hypercube=False,class_sep=10)
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y)
```

```python
class custom_perceptron_trick_upgraded:
    def __init__(self,epoch=1000,learning_rate=0.1):
        self.epoch=epoch
        self.learning_rate=learning_rate
        self.coef_=None
        self.intercept_=None

    def fit(self,x,y):
            self.coef_=np.ones(x.shape[1])
            self.intercept_=1
            
            for i in range(self.epoch):
                 temp=np.random.randint(0,x.shape[0])
                 self.coef_+=self.learning_rate*(y[temp]-(1 / (1 + np.exp(-(np.dot(x[temp], self.coef_) + self.intercept_)))))*x[temp]
                 self.intercept_+=self.learning_rate*(y[temp]-(1 / (1 + np.exp(-(np.dot(x[temp], self.coef_) + self.intercept_)))))
    
    def predict(self,x):
         return np.dot(x,self.coef_)+self.intercept_        
         
```

```python
trick=custom_perceptron_trick_upgraded()
trick.fit(x,y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
x1 = np.linspace(np.min(x[:, 0]), np.max(x[:, 0]), 100)
x2 = -(trick.coef_[0] / trick.coef_[1]) * x1 - (trick.intercept_ / trick.coef_[1])
plt.plot(x1, x2, color='red')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
class custom_perceptron_trick_upgraded_all_dataPoints:
    def __init__(self,learning_rate=0.1):
        self.learning_rate=learning_rate
        self.coef_=None
        self.intercept_=None

    def fit(self,x,y):
            self.coef_=np.ones(x.shape[1])
            self.intercept_=1
            
            for i in range(x.shape[0]):
                 self.coef_+=self.learning_rate*(y[i]-(1 / (1 + np.exp(-(np.dot(x[i], self.coef_) + self.intercept_)))))*x[i]
                 self.intercept_+=self.learning_rate*(y[i]-(1 / (1 + np.exp(-(np.dot(x[i], self.coef_) + self.intercept_)))))
    
    def predict(self,x):
         return np.dot(x,self.coef_)+self.intercept_        
         
```

```python
trick=custom_perceptron_trick_upgraded_all_dataPoints()
trick.fit(x,y)
plt.scatter(x[:, 0], x[:, 1], c=y)
x1 = np.linspace(np.min(x[:, 0]), np.max(x[:, 0]), 100)
x2 = -(trick.coef_[0] / trick.coef_[1]) * x1 - (trick.intercept_ / trick.coef_[1])
plt.plot(x1, x2, color='red')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
