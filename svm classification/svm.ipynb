```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
x,y=make_classification(n_samples=500,n_classes=2,n_features=2,n_informative=1,n_redundant=0,n_clusters_per_class=1,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3)
```

```python
X_train=pd.DataFrame(X_train)
sns.scatterplot(x=X_train[0],y=X_train[1],hue=Y_train)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',C=1)
classifier.fit(X_train,Y_train)
Y_predict=classifier.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,Y_predict))
```

```python
w = classifier.coef_[0]
b = classifier.intercept_[0]
slope = -w[0] / w[1]
intercept = -b / w[1]
y=[x*slope+intercept for x in range(-2,2)]
x=[i for i in range(-2,+2)]
```

```python
sns.scatterplot(x=X_train[0],y=X_train[1],hue=Y_train)
sns.lineplot(x=x,y=y)
```

```python
X_test=pd.DataFrame(X_test)
sns.scatterplot(x=X_test[0],y=X_test[1],hue=Y_test)
sns.lineplot(x=x,y=y)
```

```python
X_test=pd.DataFrame(X_test)
sns.scatterplot(x=X_test[0],y=X_test[1],hue=Y_predict)
sns.lineplot(x=x,y=y)
```
