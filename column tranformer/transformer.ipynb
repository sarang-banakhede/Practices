```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer=ColumnTransformer(transformers=[
    ("t1",OneHotEncoder(),["City"]),
    ("t2",OrdinalEncoder(categories=[["Male","Female"]]),["Gender"])
    ]
    ,remainder="passthrough")
```

```python
transformer=pd.DataFrame(transformer.fit_transform(df),columns=transformer.get_feature_names_out())
```
